```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
url='https://www.amazon.in/s?k=phones+under+20000&crid=1P4HT4XCLRWTW&qid=1610240223&sprefix=phones%2Caps%2C327&ref=sr_pg_1'
```

```python
req=requests.get(url)
s=BeautifulSoup(req.content,'html.parser')
Names=s.find_all('span',{'class':'a-size-medium a-color-base a-text-normal'})
Name=[]
for i in Names:
    Name.append(i.get_text())
Prices=s.find_all('span',{'class':'a-price-whole'})
Price=[]
for i in Prices:
    Price.append(i.get_text())
Ratings=s.find_all('span',{'class':'a-icon-alt'})
Rating=[]
for i in Ratings:
    Rating.append(i.get_text())
Rating=Rating[4:]
```

```python
f=pd.DataFrame({})
```

```python
f['Name']=Name
f['Price']=Price
f['Rating']=Rating
```
