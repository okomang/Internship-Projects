## Selenium Work using click way

```python
driver=webdriver.Chrome(r"C:\Users\HP\Desktop\chromedriver.exe")
driver.get('https://www.amazon.in/s?k=best+laptops&page=1&qid=1610748944&ref=sr_pg_1')
```

```python
trials=[]
```

```python
url='https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.X_pTqOgzZPY'
```

```python
req=requests.get(url)
```

```python
s=BeautifulSoup(req.content,'html.parser')
```

```python
L=s.find_all('li',{'class':'forecast-tombstone'})
```

```python
L1=L[0]
```

```python
L1.p.text
```

```python
L1.find('p',{'class':'short-desc'}).text
```

```python
Days=[]
sds=[]
for i in L:
    Days.append(i.p.text)
    sds.append(i.find('p',{'class':'short-desc'}).text)
```

```python
dess=s.find_all('div',{'class':'col-sm-10 forecast-text'})
tempall=s.find_all('p',{'class':'temp'})
```

```python
tempall
```

```python
descrip=[]
temper=[]
for i in dess:
    descrip.append(i.get_text())
for i in tempall:
    temper.append(i.get_text())
```

```python
descrip=descrip[0:7]
Days=Days[0:7]
sds=sds[0:7]
temper=temper[0:7]
```
